```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for comparison (TFR, GRR, NRR)
countries = ["Nigeria", "Japan", "USA", "India", "Germany"]
TFR = [5.0, 1.3, 1.7, 2.0, 1.5]  # Total Fertility Rate
GRR = [0.488 * tfr for tfr in TFR]  # Gross Reproduction Rate (approximate)
NRR = [grr * survival for grr, survival in zip(GRR, [0.85, 0.95, 0.98, 0.90, 0.97])]  # Net Reproduction Rate
```

```python
x = np.arange(len(countries))
width = 0.25  
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x - width, TFR, width, label="TFR", color="blue", alpha=0.75)
ax.bar(x, GRR, width, label="GRR", color="green", alpha=0.75)
ax.bar(x + width, NRR, width, label="NRR", color="red", alpha=0.75)

ax.set_xlabel("Countries")
ax.set_ylabel("Rate")
ax.set_title("Comparison of TFR, GRR, and NRR Across Countries")
ax.set_xticks(x)
ax.set_xticklabels(countries)
ax.legend()
plt.show()
```
